### Notebook to ensure that wayback archives RLLR cases

```python
import requests
import time
```

```python
# check if url is archived

def is_archived(url):

        r = requests.get('http://archive.org/wayback/available?url=' + url)
        time.sleep(2)
        if r.json()['archived_snapshots']:
            if r.json()['archived_snapshots']['closest']['status'] == '200':
                print(url, 'is archived')
                return True
        
        print(url, 'is not archived')
        return False

def save_archive(url):
    if is_archived(url):
        return None
    else:
        print("Saving...")
        r = requests.get('https://web.archive.org/save/https://' + url)
        time.sleep(2)
        
```

```python
# 2023 - 60 - 90

base_url = 'refugeelab.ca/rllr/2023rllr'

for i in range(59,91):
    url = base_url + str(i)
    #r = is_archived(url)
    r = save_archive(url)
```
